```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('Data_Train.xlsx')
df
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.drop('Additional_Info',axis=1,inplace=True)
```

```python
df
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].replace({
    'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, '4 stops':4,
})
```

```python
df['Source'].unique()
```

```python
def duration_to_minutes(duration):
    h=0
    m=0
    if 'h' in duration:
        h=int(duration.split('h')[0])
    
    if 'm' in duration:
        m=int(duration.split('m')[0].split()[-1])

    return h*60+m

df['Duration']=df['Duration'].apply(duration_to_minutes)
```

```python
df['Duration']=df['Duration']/60
```

```python
df
```

```python
df['Duration'].corr(df['Price'])
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df
```

```python
df['Dep_Time']=pd.to_datetime(df['Dep_Time'])
```

```python
df['Dep_Hour']=df['Dep_Time'].dt.hour
df
```

```python
def time_of_day(hour):
    if 5 <= hour < 11:
        return 'Morning'
    elif 11 <= hour < 16:
        return 'Noon'
    elif 16 <= hour < 22:
        return 'Evening'
    else:
        return 'Night'

df['Dep_Period'] = df['Dep_Hour'].apply(time_of_day)
```

```python
df
```

```python
df.drop('Dep_Hour',axis=1,inplace=True)
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df
```

```python
df['Date_of_Journey']=pd.to_datetime(df['Date_of_Journey'])
df['Month']=df['Date_of_Journey'].dt.month
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```
